```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df=pd.read_csv("amazon.csv")
```

```python
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '')

# 2. Convert to numeric (this handles the NaNs correctly)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# 3. Fill the missing values with the median
median_val = df['rating_count'].median()
df['rating_count'] = df['rating_count'].fillna(median_val)
```

```python



df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')


df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

df['actual_price'] = df['actual_price'].fillna(df['actual_price'].median())
df['discount_percentage'] = df['discount_percentage'].fillna(df['discount_percentage'].median())
df['discounted_price'] = df['discounted_price'].fillna(df['discounted_price'].median())
```

```python
df.isnull().sum()
```

```python
df['discount_percentage'].dtype
```

```python
df['main_category'] = df['category'].str.split('|').str[0]
```

```python
df['discount_pct'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100
```

```python

duplicate_count = df.duplicated().sum()
print(f"Number of identical rows: {duplicate_count}")


df = df.drop_duplicates()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['rating'], bins=15, kde=True, color='steelblue', edgecolor='white')
plt.title('Rating Distribution', fontsize=14)
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.savefig('rating_distribution.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['discount_percentage'], bins=15, kde=True, color='mediumpurple', edgecolor='white')
plt.title('Discount % Distribution', fontsize=14)
plt.xlabel('Discount (%)')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.savefig('discount_distribution.png', dpi=150)
plt.show()
```

```python
top_cats = df['main_category'].value_counts().nlargest(8).index
cat_disc = (df[df['main_category'].isin(top_cats)]
              .groupby('main_category')['discount_percentage']
              .mean()
              .sort_values())

plt.figure(figsize=(9, 5))
bars = plt.barh(cat_disc.index, cat_disc.values, color=sns.color_palette("Blues_d", len(cat_disc)))
plt.bar_label(bars, fmt='%.1f%%', padding=4, fontsize=10)
plt.title('Avg Discount % by Main Category (Top 8)', fontsize=14)
plt.xlabel('Avg Discount (%)')
plt.tight_layout()
plt.savefig('avg_discount_by_category.png', dpi=150)
plt.show()
```

```python
# Convert rating to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Remove NaN ratings (optional but recommended)
df = df.dropna(subset=['rating'])

cat_rating = (df[df['main_category'].isin(top_cats)]
               .groupby('main_category')['rating']
               .mean()
               .sort_values())

plt.figure(figsize=(9, 5))
bars = plt.barh(cat_rating.index, cat_rating.values, 
                color=sns.color_palette("Greens_d", len(cat_rating)))

plt.bar_label(bars, fmt='%.2f', padding=4, fontsize=10)
plt.title('Avg Rating by Main Category (Top 8)', fontsize=14)
plt.xlabel('Avg Rating')
plt.xlim(0, 5.5)
plt.tight_layout()

plt.savefig('avg_rating_by_category.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='actual_price', y='discount_percentage',
                alpha=0.6, color='steelblue', edgecolor='white', s=60)
sns.regplot(data=df, x='actual_price', y='discount_percentage',
            scatter=False, color='crimson', line_kws={'linewidth': 1.5, 'linestyle': '--'})
plt.title('Actual Price vs Discount %', fontsize=14)
plt.xlabel('Actual Price (₹)')
plt.ylabel('Discount (%)')
plt.tight_layout()
plt.savefig('price_vs_discount_scatter.png', dpi=150)
plt.show()
```

```python
bins   = [0, 1_000, 5_000, 20_000, 50_000, float('inf')]
labels = ['<1K', '1K–5K', '5K–20K', '20K–50K', '50K+']
df['rc_bucket'] = pd.cut(df['rating_count'], bins=bins, labels=labels)

bucket_counts = df['rc_bucket'].value_counts().reindex(labels)

plt.figure(figsize=(8, 5))
sns.barplot(x=bucket_counts.index, y=bucket_counts.values,
            palette='Greens_d', edgecolor='white')
for i, v in enumerate(bucket_counts.values):
    plt.text(i, v + 0.5, str(int(v)), ha='center', fontsize=10)
plt.title('Products by Rating Count Bucket', fontsize=14)
plt.xlabel('Rating Count Range')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.savefig('rating_count_buckets.png', dpi=150)
plt.show()
```
